```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
```

## Exercise 1

** Follow along with these steps: **
* ** Create a figure object called fig using plt.figure() **
* ** Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
* ** Plot (x,y) on that axes and set the labels and titles to match the plot below:**

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y)
axes.set_xlabel('x') # Notice the use of set_ to begin methods
axes.set_ylabel('y')
axes.set_title('Title')
```

## Exercise 2
**Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1]) # main axes
ax2 = fig.add_axes([0.2,.5,.2,.2]) # inset axes

```

**Now plot (x,y) on both axes. And call your figure object to show it.**

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1]) # main axes
ax2 = fig.add_axes([0.2,0.5,.2,.2]) # inset axes

# Larger Figure Axes 1
ax1.plot(x, y, 'r')
ax1.set_xlabel('x') # Notice the use of set_ to begin methods
ax1.set_ylabel('y')

# Insert Figure Axes 2
ax2.plot(x, y, 'r')
ax2.set_xlabel('x') # Notice the use of set_ to begin methods
ax2.set_ylabel('y')
```

## Exercise 3

**Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])

```

**Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
ax.plot(x,z)
ax.set_xlabel('X')
ax.set_ylabel('Z')

ax2.plot(x,y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)



fig
```

**Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

** Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
axes[0].plot(x,y,color="yellow", lw=3, ls='--')
axes[1].plot(x,z,color="green", lw=2, ls='-.')
fig
```
