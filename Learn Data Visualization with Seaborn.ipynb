```python
titanic['fare'].value_counts()
```

```python
#Draw a plot of two variables with bivariate and univariate graphs. This function provides a convenient interface to the JointGrid class, with several canned plot kinds. This is intended to be a fairly lightweight wrapper; if you need more flexibility, you should use JointGrid directly.
#join X (fare) with Y (age) in data titanic
sns.jointplot(x='fare',y='age',data=titanic ,kind='scatter')
```

```python
#Flexibly plot a univariate distribution of observations.
sns.distplot(titanic['fare'],bins=30, kde=False, color='red')
```

```python
count = titanic[(titanic['class'] == 'First' )]['age'].value_counts()
count
```

```python
#Draw a box plot to show distributions with respect to categories.
sns.boxplot(x="class", y="age", data=titanic,palette='rainbow')
```

```python
percent = titanic[(titanic['class'] == 'First' )]['age'].value_counts(normalize=True)
percent
```

```python
percent100 = titanic[(titanic['class'] == 'First' )]['age'].value_counts(normalize=True).mul(100).round(1).astype(str) + '%'
percent100
```

```python
#make new dataframe to see count, percent, and percent 100
pd.DataFrame({'count': count, 'Percent': percent, 'per100': percent100})
```

```python
#Draw a categorical scatterplot with non-overlapping points.
sns.swarmplot(x="class", y="age", data=titanic,palette='rainbow')
```

```python
titanic[(titanic['class'] == 'Third' )]['age'].value_counts()
```

```python
titanic['sex'].value_counts()
```

```python
#Show the counts of observations in each categorical bin using bars.
sns.countplot(x='sex',data=titanic)
```

```python
#Plot rectangular data as a color-encoded matrix.
sns.heatmap(titanic.corr(),cmap='coolwarm',annot=False)
plt.title('titanic.corr')
```

```python
#This class maps a dataset onto multiple axes arrayed in a grid of rows and columns that correspond to levels of variables in the dataset. The plots it produces are often called lattice, trellis, or small-multiple graphics.
g = sns.FacetGrid(titanic, col="sex")
g.map(plt.hist,'age')
```
